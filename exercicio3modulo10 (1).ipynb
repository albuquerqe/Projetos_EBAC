```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('previsao_de_renda.csv')
```

```python
variaveis_qualitativas = ['sexo', 'estado_civil', 'tipo_renda']

sns.set(style="whitegrid")

for coluna in variaveis_qualitativas:
    plt.figure(figsize=(12,6))
    sns.countplot(data=df, x='data_ref', hue=coluna)
    plt.title(f'Distribuio de {coluna} ao longo do tempo')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python

variaveis_qualitativas = ['sexo', 'estado_civil', 'tipo_renda']

for coluna in variaveis_qualitativas:
    plt.figure(figsize=(12,6))
    df_grouped = df.groupby(['data_ref', coluna])['renda'].mean().reset_index()
    sns.lineplot(data=df_grouped, x='data_ref', y='renda', hue=coluna)
    plt.title(f'Mdia de Renda ao longo do tempo por {coluna}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
