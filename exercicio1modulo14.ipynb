```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("SINASC_RO_2019.csv")

df['DTNASC'] = pd.to_datetime(df['DTNASC']) 
df['MES'] = df['DTNASC'].dt.month  

meses_interesse = {3: 'Março', 4: 'Abril', 5: 'Maio', 6: 'Junho', 12: 'Dezembro'}

for mes, nome in meses_interesse.items():
    df_mes = df[df['MES'] == mes]
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df_mes['IDADEMAE'], bins=20, kde=True)
    plt.title(f'Distribuição da idade das mães - {nome}')
    plt.xlabel('Idade da Mãe')
    plt.ylabel('Frequência')
    plt.show()
```
