```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.pointplot(x='posse_de_imovel', y='renda', data=df, errorbar='sd')
plt.title('Perfil de Média: Renda x Posse de Imóvel')
plt.ylabel('Renda Média')
plt.xlabel('Posse de Imóvel')
plt.show()
```

```python
imovel_1 = df[df['posse_de_imovel'] == 1]['renda']
imovel_0 = df[df['posse_de_imovel'] == 0]['renda']

t_stat_imovel, p_value_imovel = ttest_ind(imovel_1, imovel_0, equal_var=False)
print(f'Teste para posse de imóvel: p-value = {p_value_imovel}')
```

```python
sns.boxplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Distribuição de Renda por Posse de Veículo')
plt.show()
```

```python
veiculo_1 = df[df['posse_de_veiculo'] == 1]['renda']
veiculo_0 = df[df['posse_de_veiculo'] == 0]['renda']

t_stat_veiculo, p_value_veiculo = ttest_ind(veiculo_1, veiculo_0, equal_var=False)
print(f'Teste para posse de veículo: p-value = {p_value_veiculo}')
```
