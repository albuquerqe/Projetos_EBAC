```python
# Gráfico de barras para as variáveis categóricas: APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE
fig, axs = plt.subplots(2, 3, figsize=(18, 10))
categorical_vars = ["APGAR1", "APGAR5", "SEXO", "GESTACAO", "ESTCIVMAE"]

for i, var in enumerate(categorical_vars):
    ax = axs[i // 3, i % 3]
    sns.countplot(x=var, data=data, palette="viridis", ax=ax)
    ax.set_title(f"Distribuição de {var}")
    ax.set_ylabel("Frequência")
    ax.set_xlabel(var)

fig.delaxes(axs[1, 2])

plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras mostrando o APGAR5 médio por CONSULTAS
plt.figure(figsize=(10, 6))
sns.barplot(x="CONSULTAS", y="APGAR5", data=data, palette="coolwarm", ci=None)
plt.title("APGAR5 Médio por Número de Consultas")
plt.ylabel("APGAR5 Médio")
plt.xlabel("Número de Consultas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Painel com 4 gráficos para as variáveis SEXO, GESTACAO, ESTCIVMAE e CONSULTAS
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
panel_vars = ["SEXO", "GESTACAO", "ESTCIVMAE", "CONSULTAS"]

for i, var in enumerate(panel_vars):
    ax = axs[i // 2, i % 2]
    sns.countplot(x=var, data=data, palette="pastel", ax=ax)
    ax.set_title(f"Frequência de {var}")
    ax.set_ylabel("Frequência")
    ax.set_xlabel(var)
    ax.tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```
