```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('SINASC_RO_2019.CSV') 
```

```python
print(data.head())
```

1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 5)) 
sns.boxplot(x=data['IDADEMAE']) 
plt.title('Idade da Mãe') 
plt.xlabel('Idade da Mãe') 
plt.show()
```

2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['PESO'])
plt.title('Peso do bebê')
plt.xlabel('Peso do bebê')
plt.show()            
```

3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6)) 
sns.histplot(data['IDADEMAE'], bins=20, kde=True, color='blue') 
plt.title('Idade da Mãe') 
plt.xlabel('Idade da Mãe') 
plt.ylabel('Frequência') 
plt.show()
```

4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6)) 
sns.histplot(data['PESO'], bins=20, kde=True, color='blue') 
plt.title('Peso do Bebe') 
plt.xlabel('Peso do Bebe') 
plt.ylabel('Frequência') 
plt.show()
```
